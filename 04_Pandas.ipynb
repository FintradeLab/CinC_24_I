# **4. Cleaning**

## 4.1. Column names

```python
# Year to year
d.rename(columns = {'Year':'year'}, inplace = True)

# Country Name to country
d.rename(columns = {'Country Name':'country'}, inplace = True)

# Country Code to code
d.rename(columns = {'Country Code':'code'}, inplace = True)

# Population to pop
d.rename(columns = {'Population, total [SP.POP.TOTL]':'pop'}, inplace = True)

# Population Growth to pop_g
d.rename(columns = {'Population growth (annual %) [SP.POP.GROW]':'pop_g'}, inplace = True)

# Surface area to sur
d.rename(columns = {'Surface area (sq. km) [AG.SRF.TOTL.K2]':'sur'}, inplace = True)

# Life expectancy to lexp
d.rename(columns = {'Life expectancy at birth, total (years) [SP.DYN.LE00.IN]':'lexp'}, inplace = True)

# Net Migration to net
d.rename(columns = {'Net migration [SM.POP.NETM]':'net'}, inplace = True)
```

# **8. Graphs**

```python
# Library
import plotly.express as px
```

## 8.1. Line Graph

```python
france = d2.query("country ==  'France'")
france.head()
```

```python
# Fig 1
fig1 = px.line(france, x = 'year_str', y = 'pop_g',
               title="Population growth of France")
fig1.show()
```

```python
# Fig 1
fig2 = px.line(france, x = 'year_str', y = 'pop_g')

# title

fig2.update_layout(title_text = 'Population growth of France',
                   title_font_size = 30,
                   title_x = 0.5)

# axis
fig2.update_xaxes(title = 'Year', title_font_size = 20)
fig2.update_yaxes(title = 'Percentage (%)', title_font_size = 20)

fig2.show()
```

## 8.2. Bar graph

```python
# Fig 3
fig3 = px.bar(france, x = 'year_str', y = 'pop_g')

# title

fig3.update_layout(title_text = 'Population growth of France',
                   title_font_size = 30,
                   title_x = 0.5)

# axis
fig3.update_xaxes(title = 'Year', title_font_size = 20)
fig3.update_yaxes(title = 'Percentage (%)', title_font_size = 20)

fig3.show()
```

## 8.3. More than one line

```python
## Create a list of countries
c_list = ['Colombia', 'Chile', 'Peru', 'Mxico', 'Brazil', 'Argentina']

## Extract the countries from the dataframe
### First we must unindexed the df
d3 = d2.reset_index()
latam = d3[d3['country'].isin(c_list)]

latam.head()
```

```python
fig4 = px.line(latam, x = 'year_str', y = 'pop_g', color = 'country')

# title

fig4.update_layout(title_text = 'Population growth of Latam',
                   title_font_size = 30,
                   title_x = 0.5)

# axis
fig4.update_xaxes(title = 'Year', title_font_size = 20)
fig4.update_yaxes(title = 'Percentage (%)', title_font_size = 20)

fig4.show()
```

## 8.4. Pie chart

```python
# The idea is to graph the top 20 countries
# by population and the year 2020

y2020_2 = y2020.sort_values('pop', ascending = False)
y2020_2.head()

## Top 20 by population
y2020_2 = y2020_2.iloc[0:20]
```

```python
fig5 = px.pie(y2020_2, values = 'pop', names = y2020_2.index.get_level_values(0))

# title
fig5.update_layout(title_text = 'Top 20 Countries by Population - 2020',
                   title_font_size = 30,
                   title_x = 0.5)
fig5.show()
```
